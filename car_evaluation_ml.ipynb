Car Evaluation Machine Learning Model using K-Nearest Neighbours Algorithm 

1)Importing all the models required for the model


```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing
```

2) Reading in the dataset using Pandas


```python
data =pd.read_csv("car.data")
print(data.head())
```

3) Encoding string value attibutes to numerical values

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))

print(buying)
```

```python
predict = "class"
```

4) Assigning features to the variables and splitting into training and testing data



```python
X = list(zip(buying, maint, door, persons, lug_boot, safety))
y = list(cls)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

print(x_train, y_test)
```

5) Fit model with training data and test accuracy using testing data


```python
model = KNeighborsClassifier(7)

model.fit(x_train, y_train)

```

```python
acc= model.score(x_test, y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
```

```python
names = ["unacc", "acc", "good", "vgood"]
```

```python
for x in range(len(x_test)):
  print("predicted:" , names[predicted [x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
  n = model.kneighbors([x_test[x]], 9,True)
  print("N: ", n)
```
